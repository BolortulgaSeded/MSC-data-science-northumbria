```python
import pandas as pd
import numpy as np
df = pd.read_csv('Placement_Data_Full_Class.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(df['salary'])
plt.show()

sns.histplot(df['salary'], kde=True)
plt.show()
```

```python
df_mean = df.copy()
df_mean['salary'] = df_mean['salary'].fillna(df_mean['salary'].mean())
print("Mean used:", df['salary'].mean())
print("Filled with mean.new null count:",df_mean['salary'].isnull().sum())
```

```python
df_median = df.copy()
df_median['salary'] = df_median['salary'].fillna(df_median['salary'].mean())
print("Median used:", df['salary'].median())
print("Filled with median.new null count:",df_median['salary'].isnull().sum())
```

```python
df_mode = df.copy()
mode_val = df_mode['salary'].mode()[0]
df_mode['salary'] = df_mode['salary'].fillna(mode_val)
print(f"Filled with Mode ({mode_val}). New null count:", df_mode['salary'].isnull().sum())
```
